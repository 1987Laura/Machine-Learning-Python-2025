# matplotlib = matematica + grafica + librarie


```python
import matplotlib.pyplot as plt
```

```python
plt.scatter([1,2],[2,3])
plt.show()
```

```python
plt.plot([1,1],[1,2])
plt.show()
```

```python
plt.bar([1,2.5,4], [2,3,5])
plt.show()
```

```python
plt.barh([1,2.5,4], [2,3,5])
plt.show()
```

```python
plt.pie([1,2.5,4])
plt.show()
```

```python
import numpy as np
```

```python
x =np.arange(5)
y = np.arange(5) * 2
plt.scatter(x,y)
plt.plot(x,y)
plt.bar(x,y, color="green")
```

```python
x =np.linspace(2,8,5)
y = np.linspace(2,8,5) * 2
plt.scatter(x,y)
plt.plot(x,y)
plt.bar(x,y, color="green")
```

```python
x =np.random.randint(2,8,5)
y = np.random.randint(2,8,5) * 2
plt.scatter(x,y)
plt.plot(x,y)
```
