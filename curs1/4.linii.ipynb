```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def f(x):
    a = 4
    b = 2
    return a * x + b 

suprafete = np.arange(40,51)
preturi = np.array(list(map(f, suprafete)))

plt.scatter(suprafete,preturi)
plt.plot(suprafete,preturi)

```

```python
def f(x):
    a = -1
    b = 2
    return a * x + b 

suprafete = np.arange(40,51)


preturi = np.array(list(map(f,suprafete)))

plt.scatter(suprafete, preturi)
plt.plot(suprafete, preturi);

```
