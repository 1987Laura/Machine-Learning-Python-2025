```python
from scipy import stats
```

```python
rezultat = stats.linregress(suprafete,preturi)
rezultat
```

```python
def f_perfect(x):
    a = rezultat.slope
    b = rezultat.intercept
    return a * x + b

preturi_generate_cu_f_perfect = np.array(list(map(f_perfect,suprafete)) )

plt.plot(suprafete, preturi_generate_cu_f_perfect, c ="red")


def f1(x):
    a=2.2
    b=5
    return a * x + b

preturi_generate_cu_f1 = np.array(list(map(f1,suprafete)) )
preturi_generate_cu_f1

plt.plot(suprafete,preturi)
plt.plot(suprafete, preturi_generate_cu_f1)
plt.scatter(suprafete,preturi)


def f2(x):
    a= 1.9
    b= -2
    return a * x + b

preturi_generate_cu_f2 = np.array(list(map(f2,suprafete)) )
preturi_generate_cu_f2

plt.plot(suprafete,preturi, c="green")
plt.plot(suprafete, preturi_generate_cu_f2, c= "lime")
plt.scatter(suprafete,preturi)
```
